```python
# Importe a biblioteca pandas
import pandas as pd
import numpy as np
```

```python
# Crie uma Serie contendo os seguintes elementos [4, 7, -5, 3]
obj = pd.Series([4, 7, -5, 3])
obj
```

```python
# Crie um objeto Series contendo [4, 7, -5, 3] e 
#   defina seu próprio array de índices como sendo ['d', 'b', 'a', 'c']
obj2 = pd.Series([4, 7, -5, 3, 'a'], index=['d', 'b', 'a', 'c', 'f'])
obj2
```

```python
# Crie uma série como seguinte dicionário 
# {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
obj3 = pd.Series(sdata)
obj3
```

```python
# Crie uma série como seguinte dicionário 
# sdata= {'Ohio': 35000, 'Texas': 71000, 'Oregon': None, 'Utah': None}
# mas especificando como índice este vetor ['California', 'Ohio', 'Oregon', 'Texas']
states = ['California', 'Ohio', 'Oregon', 'Texas']
obj4 = pd.Series(sdata, index=states)
obj4
```

```python
# Operações com indexação automática simplificam a sintaxe e a lógica de programação
# Some as séries criadas nos dois exercícios anteriores (obj3 e obj4)
obj3 + obj4
```

```python
# Crie um data frame a partir dos dados do seguinte dicionário, chame-o de frame
data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
'year': [2000, 2001, 2002, 2001, 2002, 2003],
'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}
frame = pd.DataFrame(data)
frame
```

```python
# Acesse a coluna state usando a sintaxe com colchetes
frame['state']
```

```python
# Atribua o valor 16.5 para todas as posições da coluna debt
# Quando a coluna não existir, o pandas cria ela pra você
frame['debt'] = 16.5
frame
```

```python
# Atribua o valor de um intervalo de 0 a 5 (float) na coluna debt
frame['debt'] = np.arange(6.)
frame
```

```python
# Crie uma nova coluna chamada eastern com o valor True para 
#   as linhas cujo estado seja Ohio, e False caso contrário
frame['eastern'] = frame.state == 'Ohio'
frame
```

```python
diamonds = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv?raw=true')
diamonds
```

```python
regra = (diamonds['x'] > 6) & (diamonds['y']>6) & (diamonds['z']>6)
```

```python
diamonds[regra]
```

```python
diamonds.loc[regra, ['price', 'x', 'y']]
```

```python
# dois pontos sem valor retorna todos
diamonds.loc[:, ['price', 'x', 'y']]
```

```python
# dois pontos com valor retorna o intervalo de linhas que deseja
diamonds.loc[50:64, ['price', 'x', 'y']]
```

```python
# iloc retorna as linhas e colunas que escolher 
diamonds.iloc[[9,10, 11, 12], [6, 7, 8, 9]]
```

```python

# Considere os seguintes DataFrames
df_medico = pd.DataFrame({'cpf': ['11222731642','91498733332','81464221612'], 
                    'nome': ['Jose Pereira', 'Maria da Silva', 'Pedro Martins'], 
                    'Salario': [10, 20, 15]})
df_professor = pd.DataFrame({'cpf': ['11222731642','91498733332','21564281600'], 
                    'nome': ['Jose Pereira', 'Maria da Silva', 'Roberto Afonso'], 
                    'Salario': [6, 8, 5]})
```

```python
df_medico.head()
```

```python
df_professor.head()
```

```python
# Faça o merge dos dois dataframes usando como chave de junção a coluna 'key'
# O parametro on neste caso é redundante. Funcionaria sem ele.
pd.merge(df_medico, df_professor, on='cpf', suffixes=['_M','_P'])

#retorna somente os que estão na intersecção das duas tabelas
```

```python
df1 = pd.merge(df_medico, df_professor, on = ['cpf'], how = 'outer', validate ='m:m')
df1.head()
```

```python
df1.fillna(" ", inplace = True)
df1
```
